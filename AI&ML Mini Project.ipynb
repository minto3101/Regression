```python
import pandas as pd
import seaborn as sns
```

```python
Xdf = pd.read_csv("exercise.csv")
Xdf
```

```python
Ydf = pd.read_csv("calories.csv")
Ydf
```

```python
#combining the two datasets
calories_data = pd.concat([Xdf, Ydf['Calories']], axis=1)
calories_data
```

```python
#Data Preprocessing
#Converting categorical variables into numerical variables 
calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
calories_data
```

```python
#dropping useless columns
calories_data.drop(columns=["User_ID"], inplace = True)
calories_data
```

```python
X = calories_data.iloc[:,0:7]
X
```

```python
y = calories_data.iloc[:,-1]
y
```

```python
#splitting dataset into testing and training dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train
```

```python
y_train
```

```python
#creating linear regression model
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
#finding mean absolute error for model 1
y_pred = model1.predict(X_test)
y_pred
```

```python
from sklearn import metrics
mae1 = metrics.mean_absolute_error(y_test, y_pred)
mae1
```

```python
#creating RandomForestRegression model
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor(n_estimators=100)
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

```python
#finding mean absolute error for model 2
y_pred = model2.predict(X_test)
print(y_pred)
mae2 = metrics.mean_absolute_error(y_test, y_pred)
mae2
```
